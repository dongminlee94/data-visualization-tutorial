## 기본 사용법

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Plotting

- pyplot 객체를 사용하여 데이터를 표시
- pyplot 객체에 그래프들을 쌓은 다음 show로 flush
- 그래프는 원래 figure 객체에 생성됨. figure는 우리가 그래프를 그릴 수 있는 모판이라고 생각하면 됨.
- pyplot 객체 사용시, 기본 figure에 그래프가 그려짐

**Example 1**

```python
X = range(100)
Y = range(100)
```

```python
plt.plot(X, Y)
plt.show()
```

**Example 2**

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X]
```

```python
X_2 = range(100)
Y_2 = [np.sin(value) for value in X]
```

```python
plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

**두개의 plot 생성**

- 전체 figure 안에 sub-figure를 둘 수도 있다.

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10,5) # 크기지정

ax_1 = fig.add_subplot(1,2,1) # 두개의 plot 생성
ax_2 = fig.add_subplot(1,2,2)  # 두개의 plot 생성

ax_1.plot(X_1, Y_1, c="b")  # 첫번째 plot
ax_2.plot(X_2, Y_2, c="g")  # 두번째 plot
plt.show()
```

### Set color

- Color 속성을 사용
- Float -> 흑백, rgb color, predefined color 사용(ex. r - red, g - green, b - blue)
- http://matplotlib.org/2.0.2/api/colors_api.html

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]
```

```python
plt.plot(X_1, Y_1, color="#2EFE2E")
plt.plot(X_2, Y_2, color="r")
plt.show()
```

### Set linstyle

- ls 또는 linestyle 속성 사용
- https://matplotlib.org/examples/lines_bars_and_markers/linestyles.html

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed") # c = color
plt.plot(X_2, Y_2, c="r", ls="dotted")
plt.show()
```

### Set title

- Pyplot에 title함수 사용, figure의 subplot별 입력 가능

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")

plt.title("Two lines")
plt.show()
```

**Latex 타입의 표현도 가능 (수식 표현 가능)**

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")

plt.title('$y = \\frac{ax + b}{test}$')
plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")

plt.title('$y = ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')

plt.show()
```

### Set legend

- Legend 함수로 범례를 표시함, loc 위치등 속성 지정

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=True, loc="upper right") # loc = location - "lower right" 도 가능

plt.title('$y = ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')

plt.show()
```

### Set grid & xylim

- Graph 보조선을 긋는 grid와 xy축 범위 한계를 지정

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=True, loc="lower right")

plt.grid(True, lw=0.4, ls="--", c=".90")
plt.xlim(-100, 200)
plt.ylim(-200, 200)

plt.show()
```

## Matplotlib Graph

### Scatter

- scatter 함수 사용, marker: scatter 모양지정

```python
data_1 = np.random.rand(512, 2) # random samples from a uniform distribution over [0, 1)
data_2 = np.random.rand(512, 2)
```

```python
plt.scatter(data_1[:,0], data_1[:,1], c="b", marker="x")
plt.scatter(data_2[:,0], data_2[:,1], c="r", marker="o")

plt.show()
```

- s: 데이터의 크기를 지정, 데이터의 크기 비교 가능

```python
N = 50

x = np.random.rand(N)
y = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2 
colors = np.random.rand(N)

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

area의 값에 따라서 s에 크기를 지정하여 비교할 수 있다.

### Bar chart

- Bar 함수 사용

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]

X = np.arange(0,8,2) # [0 2 4 6]
```

```python
plt.bar(X + 0.0, data[0], color = 'b', width = 0.5)
plt.bar(X + 0.5, data[1], color = 'g', width = 0.5)
plt.bar(X + 1.0, data[2], color = 'r', width = 0.5)
plt.xticks(X + 0.5, ("A","B","C","D"))
plt.show()
```

```python
data = np.array([[5., 25., 50., 20.],
                 [4., 23., 51., 17],
                 [6., 22., 52., 19]])
color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])
```

```python
for i in range(data.shape[0]):
    plt.bar(X, data[i], bottom=np.sum(data[:i], axis=0), 
            color=color_list[i], label=data_label[i])
plt.legend()
plt.show()
```

### Histogram

```python
X = np.random.randn(100) 
# random floats sampled from a univariate “normal” (Gaussian) distribution of mean 0 and variance 1

plt.hist(X, bins=100) # bins - 데이터를 나누는 개수
plt.show()
```

### Boxplot

```python
data = np.random.randn(100,5)

plt.boxplot(data)
plt.show()
```

출력된 그림을 보면 가운데 네모칸으로 되어있는 칸에 50%의 데이터가 있고, 위아래로 나머지 데이터들이 분포되어있다고 보면 된다. 네모칸 안에 노란색은 평균값을 말해준다. 양 끝에 보면 o 라는 것도 시각화가 되어있는데 outlier를 표현해주는 것이다.

## Matplotlib with pandas

### Pandas matplotlib

- Pandas 0.7 버전이상 부터 matplotlib를 사용한 그래프 지원 
- Dataframe, series별로 그래프 작성 가능

- 데이터간의 상관관계를 볼 때 scatter graph 사용 가능

```python
import pandas as pd
```

```python
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data' # Data URL
df_data = pd.read_csv(data_url, sep='\s+', header = None) 
df_data.columns = [
    'CRIM','ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO' ,'B', 'LSTAT', 'MEDV'] 
# Column Header 이름 지정
```

```python
df_data.head()
```

```python
fig = plt.figure() # figure 반환
ax = []

for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i)) # 4개의 plot 생성
    
ax[0].scatter(df_data["CRIM"], df_data["MEDV"])
ax[1].scatter(df_data["PTRATIO"], df_data["MEDV"])
ax[2].scatter(df_data["AGE"], df_data["MEDV"])
ax[3].scatter(df_data["NOX"], df_data["MEDV"])
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10.0, 10.0, forward=True)
ax = []

for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i))

columns = ["CRIM", "PTRATIO", "AGE", "NOX"]
colors = ["b", "g", "c", "r"]

for i in range(4):
    ax[i].scatter(df_data[columns[i]], df_data["MEDV"], color=colors[i], label=columns[i])
    ax[i].legend()
    ax[i].set_title(columns[i])

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

- plot 함수를 사용하면 전체 데이터의 graph를 그림

```python
df_data.plot()
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10,5)

ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(df_data["MEDV"])
ax_2.hist(df_data["MEDV"], bins=50)

ax_1.set_title("House price MEDV")
ax_2.set_title("House price MEDV")

plt.show()
```

```python
pd.plotting.scatter_matrix(df_data, diagonal="kde", alpha=1,  figsize=(20, 20))
plt.show()
```

가운데는 데이터들이 어떠한 분포를 가지고 있는지를 보여주고, 나머지들은 각각의 데이터들이 어떠한 상관관계를 가지고 있는지를 보여준다.
